```python
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
font = {"size": 10, "weight": "bold"}
matplotlib.rc("font", **font)
```

```python
import imp

plot = imp.load_source('plot', '../src/plot.py')
```

```python
color_c = matplotlib.cm.viridis(0.4)
```

```python
for tournament_type in ['standard', 'noise', 'probend', 'probend_noise']:
    df = pd.read_csv('../output/%s_importance.csv' % tournament_type, index_col=0)
    plt.figure()
    plt.title("Feature importances")
    plt.bar(
        range(len(df['feature'])),
        df['feature_importance'],
        color=color_c,
        yerr=df['std'],
        align="center",
    )

    xticks = [plot.features_labels[f] for f in df.feature]

    plt.xticks(range(len(df['feature'])), xticks, rotation=90);
    plt.xlim([-1, len(df['feature'])]);
    plt.savefig(
        "../images/%s_feature_importance_bar_plot.pdf" % tournament_type,
        bbox_inches="tight",
    )
```
