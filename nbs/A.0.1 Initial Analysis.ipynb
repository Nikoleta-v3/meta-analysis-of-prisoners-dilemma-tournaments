```python
fig, axes = plt.subplots(figsize=(20, 10))
axes.hist(df['Cooperation_rating'], bins=20); #axes.set_xlabel('Cooperation Rating', fontsize=25);
axes.set_ylabel('Frenquency', fontsize=25)
plt.tight_layout()
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#fig.savefig('/home/nightwing/rsc/Annual-Report/img/cooperating_ratio.pdf')
plt.show()
```

Compaired to the previous data, there seems to be more comperative enviroments. We see a high peek at 0.5.

```python
fig, axes = plt.subplots(figsize=(20, 10))
df.groupby(['size']).mean()['Cooperation_rating'].plot();
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

fig.suptitle('States', fontsize=25)
axes[0][0].hist(df['CC_rate'], bins=20); axes[0][0].set_xlabel('CC state', fontsize=25)
axes[0][1].hist(df['CD_rate'], bins=20); axes[0][1].set_xlabel('CD state', fontsize=25)

axes[1][0].hist(df['DC_rate'], bins=20); axes[1][0].set_xlabel('DC state', fontsize=25)
axes[1][1].hist(df['DD_rate'], bins=20); axes[1][1].set_xlabel('DD state', fontsize=25)
plt.tight_layout()
#fig.savefig('/home/nightwing/rsc/Annual-Report/img/states_hist.pdf')
plt.show()
```

Looking at the data the following stand out:

- Cooperative enviroments. It is seems that more that the 50% of the tournaments have had a cooperation rating
greater than 0.5

- The most time spend in each of the above states (CC, CD, DC, DD) is in CC
