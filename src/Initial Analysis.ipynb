Initial Analysis
----------------

A few pretty plots.

```python
import matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv('data_frame.csv')
```

```python
df.columns
```

Rates for turns and Prob ends
------------------------------

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 10))
for i, col in enumerate(['turns', 'probend']):
    dummy = df[df[col].notnull()]
    axes[i,0].scatter(dummy[col], dummy['CC_rate']); axes[i,0].set_title('CC_rate {}'.format(col));
    axes[i,1].scatter(dummy[col], dummy['CD_rate']); axes[i,1].set_title('CD_rate {}'.format(col));
    axes[i,2].scatter(dummy[col], dummy['DC_rate']); axes[i,2].set_title('DC_rate {}'.format(col));
    axes[i,3].scatter(dummy[col], dummy['DD_rate']); axes[i,3].set_title('DD_rate {}'.format(col));
    axes[i,4].scatter(dummy[col], dummy['Cooperation_rating']); axes[i,4].set_title('Cooperation_rating {}'.format(col));
```

Rates for seed, noise, repetitions
------------------------------

```python
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(25, 15))
for i, col in enumerate(['seed', 'noise', 'repetitions', 'size']):

    axes[i,0].scatter(df[col], df['CC_rate']); axes[i,0].set_title('CC_rate {}'.format(col));
    axes[i,1].scatter(df[col], df['CD_rate']); axes[i,1].set_title('CD_rate {}'.format(col));
    axes[i,2].scatter(df[col], df['DC_rate']); axes[i,2].set_title('DC_rate {}'.format(col));
    axes[i,3].scatter(df[col], df['DD_rate']); axes[i,3].set_title('DD_rate {}'.format(col));
    axes[i,4].scatter(df[col], df['Cooperation_rating']); axes[i,4].set_title('Cooperation_rating {}'.format(col));
```

Cooperation_rating
-----------------

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
fig.suptitle('Cooperation Rating')
axes[0].scatter(df['Rank'], df['Cooperation_rating']); axes[0].set_title('Rank');
axes[1].scatter(df['Median_score'], df['Cooperation_rating']); axes[1].set_title('Median_score');
axes[2].scatter(df['Wins'], df['Cooperation_rating']); axes[2].set_title('Wins');
```
