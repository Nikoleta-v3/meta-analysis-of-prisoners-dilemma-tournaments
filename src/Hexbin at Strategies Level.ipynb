Hexbin at Strategies Level
---------------------------

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('df.csv')
```

```python
def hexbin_plot(x, y, i, j=False,):
    """
    A function which takes an x and y and returns a hexbin plot for a given axe.
    """
    x = x
    y = y
    xmin = x.min()
    xmax = x.max()
    ymin = y.min()
    ymax = y.max()
    
    if j is False:
        ax = axes[i]
    else:
        ax = axes[i][j]
    

    hb = ax.hexbin(x, y, gridsize=50, cmap='plasma')
    ax.axis([xmin, xmax, ymin, ymax])

    cb = fig.colorbar(hb, ax=ax)
    cb.set_label('counts')
```

Turns, noise, and prob ends
------------------------------

```python
fig, axes = plt.subplots(nrows=3, ncols=5, sharey=True, figsize=(25, 15))
for i, col in enumerate(['turns', 'noise', 'probend']):
    dummy = df[df[col].notnull()]
    
    hexbin_plot(dummy[col], dummy['CC_rate'], i, 0);axes[i,0].set_title('CC_rate {}'.format(col))
    hexbin_plot(dummy[col], dummy['CD_rate'], i, 1);axes[i,1].set_title('CD_rate {}'.format(col))
    hexbin_plot(dummy[col], dummy['DC_rate'], i, 2);axes[i,2].set_title('DC_rate {}'.format(col))
    hexbin_plot(dummy[col], dummy['DD_rate'], i, 3);axes[i,3].set_title('DD_rate {}'.format(col)); 
    hexbin_plot(dummy[col], dummy['Cooperation_rating'], i, 4);axes[i,4].set_title('Cooperation_rating {}'.format(col))
            
# hb = ax.hexbin(x, y, gridsize=50, bins='log', cmap='inferno')
```

Rates for seed, size, repetitions
------------------------------

```python
fig, axes = plt.subplots(nrows=3, ncols=5, sharey=True, figsize=(25, 15))
for i, col in enumerate(['seed', 'repetitions', 'size']):
    
    hexbin_plot( df[col], df['CC_rate'], i, 0);axes[i,0].set_title('CC_rate {}'.format(col))
    hexbin_plot(df[col], df['CD_rate'], i, 1);axes[i,1].set_title('CD_rate {}'.format(col))
    hexbin_plot(df[col], df['DC_rate'], i, 2);axes[i,2].set_title('DC_rate {}'.format(col))
    hexbin_plot(df[col], df['DD_rate'], i, 3);axes[i,3].set_title('DD_rate {}'.format(col)); 
    hexbin_plot(df[col], df['Cooperation_rating'], i, 4);axes[i,4].set_title('Cooperation_rating {}'.format(col))
            
# hb = ax.hexbin(x, y, gridsize=50, bins='log', cmap='inferno')
```

Cooperation rating
-----------------

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 10))
fig.suptitle('Cooperation Rating')
hexbin_plot(df['Rank'], df['Cooperation_rating'], 0); axes[0].set_title('Median_score');
hexbin_plot(df['Median_score'], df['Cooperation_rating'], 1); axes[1].set_title('Median_score');
hexbin_plot(df['Wins'], df['Cooperation_rating'], 2); axes[2].set_title('Wins');
```
